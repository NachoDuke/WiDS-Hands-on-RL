```python
#name of the input file
# filename = 'mdp-2-2.txt'
filename = 'mdp-10-5.txt'
with open(filename,'r') as f:
    contents = f.readlines()
for line in contents:
    words = line.split(' ')
    if words[0]=='states':
        numstates = int(words[1])
        #actions will contain tuples containing action, next state, reward and probability
    elif words[0]=='actions':
        numactions = int(words[1])
        actions = [[[] for i in range(numactions)] for i in range(numstates)]
    elif words[0]=='gamma':
#         words = line.split('  ')
        gamma = float(words[2].lstrip())
#         print(words[2])
    else:
        actions[int(words[1])][int(words[2])].append((int(words[3]),float(words[4]),float(words[5])))

print(numstates,actions,numactions)
```

```python
margin = pow(10,-8)
valfunc = [0 for i in range(numstates)]
pi = [0 for i in range(numstates)]

while True:
    while True:
        error = 0
        for state in range(numstates):
            oldV = valfunc[state]
            for action in range(numactions):
                val = 0
                for s0 in range(numstates):
                    tups = actions[state][action]
                    for tup in tups:
                        if tup[0]==s0:
                            break
                    else:
                        continue
#                     print(tup[0],tup[1],tup[2])
                    val+=tup[2]*(tup[1] + gamma*valfunc[s0])
                valfunc[state] = max(valfunc[state],val)
            error = max(error,abs(oldV - valfunc[state]))
        if error < margin:
            break
#     print(valfunc)
    same = True
    for state in range(numstates):
        oldpi = pi[state]
        a_max = 0
        v_max = 0
        for action in range(numactions):
            val = 0
            for s0 in range(numstates):
                tups = actions[state][action]
                for tup in tups:
                    if tup[0]==s0:
                        break
                else:
#                     print()
                    continue
                val = val + tup[2]*(tup[1] + gamma*valfunc[s0])
            if val > v_max:
                a_max = action
                v_max = val
        pi[state] = a_max
        if pi[state] != oldpi:
            same = False
    if same:
        break
with open('answerpolicy'+filename,'w') as f:
    for i in range(numstates):
        f.write("{:.6f}".format(valfunc[i])+' '+str(pi[i])+'\n')
```
